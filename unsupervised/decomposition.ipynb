```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/classfication/dataset_185_yeast.csv')
df
```

```python
df['class_protein_localization'].value_counts()
```

```python
X = df.drop('class_protein_localization', axis=1)
y = df['class_protein_localization']
```

```python
dec = PCA(n_components=3)
X_decomposed = dec.fit_transform(X)
df_x = pd.DataFrame(X_decomposed, columns=['PC1', 'PC2', 'PC3'])
df_x['class_protein_localization'] = y
```

```python
px.scatter(df_x, x='PC1', y='PC2', color='class_protein_localization')
```

```python
px.scatter_3d(df_x, x='PC1', y='PC2', z='PC3', color='class_protein_localization')
```

```python
px.treemap(df_x, path=['class_protein_localization'], values='PC1')
```
