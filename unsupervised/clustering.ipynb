```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.cluster import KMeans, MiniBatchKMeans, BisectingKMeans
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
px.scatter(df, x='x', y='y')
```

```python
output = []
for i in range(1, 11):
    obj = KMeans(n_clusters=i, n_init='auto').fit(df[['x', 'y']])
    output.append({
        'wcss': obj.inertia_,
        'cluster': i,
        'algo': obj
    })
```

```python
data = pd.DataFrame(output)
px.line(data, x='cluster', y='wcss', title='Elbow Method')
```

```python
df['groups'] = output[2]['algo'].labels_
```

```python
px.scatter(df, x='x', y='y', color='groups', symbol='groups')
```
